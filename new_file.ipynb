```python
df['age'] = df['age'].apply(lambda x: np.nan if not str(x).isnumeric() else int(x))

df['age'].apply(lambda x: np.nan if x > 100 else x)
```

### 6. clean sal by removing curr sysmbols and converting it to numeric

```python
df
```

```python
df['salary'].dtypes
```

```python
df['salary'] = df['salary'].astype(str)
```

```python
# df['salary'] = df['salary'].apply(lambda x : x if isnumeric() else np.nan if x.lower() == "nan" else )
```

```python
df['salary'] = df['salary'].str.replace("₹", "",regex=False).replace("N/A",np.nan)
df['salary'] = pd.to_numeric(df['salary'], errors = "coerce")
```

```python
import datetime
df['join_date'] = pd.to_datetime(df["join_date"], errors = 'coerce',dayfirst=True)
```

### standardize department values by trimming space and converting to uppercase

```python
df['department'].unique()
```

```python
df['department'] = df['department'].str.replace(" ","")
df['department'] = df['department'].str.upper()
df
```

### 10 Normalize city names to a consistent case spelling

### 11. convert performance score into numeric by mapping text valuies to numbers.


```python
df['performance_score'].unique()
```

```python
df['performance_score'] = df['performance_score'].replace(['excellent','poor'], ['5','1'])
df['performance_score'] = pd.to_numeric(df['performance_score'],errors='coerce')
df['performance_score']
```

### 12. clean exp yrs by converting text values to numeric and removing negatives

```python
df
```

```python
df['experience_years'].unique()
```

```python
df['experience_years'] = df['experience_years'].replace(['five','-1'],['5',np.nan])
df['experience_years'] = pd.to_numeric(df['experience_years'],errors='coerce')
df['experience_years']
```

### 13. convert is active values (yes/no/true/false) in to boolean format

```python
df['is_active'].unique()
```

```python
df['is_active'] = df['is_active'].replace(['yes','no'],[True,False])
df['is_active'].unique()
```
